```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
import warnings
warnings.filterwarnings('ignore')

import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats

%matplotlib inline

house = pd.read_csv('./kc_house_data.csv', parse_dates = ['date'])

house.columns
house['price'].describe()
#inspect if any data is missing
print(house.isnull().any())
# Inspecting type
print(house.dtypes)
#convert the types of data
house['waterfront'] = house['waterfront'].astype('category',ordered=True)
house['view'] = house['view'].astype('category',ordered=True)
house['condition'] = house['condition'].astype('category',ordered=True)
house['grade'] = house['grade'].astype('category',ordered=False)
house['zipcode'] = house['zipcode'].astype(str)
house['id'] = house['id'].astype(str)
house.head(2)
house.shape
```

```python
#Looking summary statistics
house.shape
house.describe(include='all')
#drop outlier, confirmed error on Kaggle's discussion
house.drop(house[house.bedrooms == 33].index, inplace = True)
house.shape
```
